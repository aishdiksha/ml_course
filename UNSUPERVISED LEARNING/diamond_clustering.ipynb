```python
import pandas as pd
import plotly.express as  px
from sklearn.cluster import KMeans
import plotly.io as pio
pio.renderers.default = "kaggle"
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/diamonds.csv')
df.head()
```

```python
px.scatter_3d(df, x='carat', y='table', z='depth').show()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
X = df[['depth','table','carat']]
X
```

```python
wcss = []
for i in range(1,11):
    kmeans =KMeans(n_clusters=i,random_state=0).fit(X)
    print(f'{i} cluster:{kmeans.inertia_}')
    wcss.append(kmeans.inertia_)
px.line(y=wcss,x=range(1,11),title="WCSS")
```

```python
kmeans = KMeans(n_clusters=5,random_state=0).fit(X)
df['group'] = kmeans.predict(X)
```
